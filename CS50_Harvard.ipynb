```python
#Return
def main():
    x = int(input('Enter the number'))
    print('X squared is', square(x))
    print('X squared3 is', square3(x))
    print('X plus X = ', sum(x))
    print('X minus X = ', devide(x))
    
def square(n):
    return n * n

def square3(n):
    return pow(n,3)

def sum(n):
    return n + n

def devide(n):
    return n- n

main()
```

```python
#Mario
def main():
    print_row(4)
    print_col(3)
    print_square(3)
    
def print_row(width):
    print("?" * width)

def print_col(height):
    print("#\n" * height, end="")
    
def print_square(size):
    for i in range(size):
        for j in range(size):
            print("#", end="")
        print()
     
main()    
```
