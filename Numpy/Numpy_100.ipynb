44. Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates (★★☆)

```python
cart_matrix = np.random.random([10,2])
x = cart_matrix[:, 0]
y = cart_matrix[:, 1]

r = np.sqrt(x**2, y**2)
theta = np.arctan(y, x)

polar = np.stack((r, theta), axis=1)

print("Cartesian Coordinates:\n", cart_matrix)
print("\nPolar Coordinates (r, theta):\n", polar)
```

46. Create a structured array with x and y coordinates covering the [0,1]x[0,1] area (★★☆)

```python
n = 10

x = np.linspace(0, 1, n)
y = np.linspace(0, 1, n)

xx, yy = np.meshgrid(x, y)

structured = np.zeros(n*n, dtype=[('x', float), ('y', float)])
structured['x'] = xx.ravel()
structured['y'] = yy.ravel()

print(structured)
```

48. Print the minimum and maximum representable values for each numpy scalar type (★★☆)

```python
int_types = [int8, int16, int32, int64, uint8, uint16, uint32,uint64]
for i in int_types:
    info = np.iinfo(i)
    print(f'{i.__name__}: min: {info.min}, max: {info.max}')
    
float = [float16, float32, float64]
for f in float:
    info = np.finfo(f)
    print(f'{f.__name__}: min: {info.min}, max: {info.max}')
```

```python
dtype = [('position', [('x', float16), ('y', float16)]), ('color', [('r', int), ('g', int), ('b', int)])]
arr = np.zeros(3, dtype=dtype)
arr['position']['x'] = [1.0, 2.0, 3.0]
arr['position']['y'] = [4.0, 5.0, 6.0]
arr['color']['r'] = [255, 128, 0]
arr['color']['g'] = [0, 255, 128]
arr['color']['b'] = [128, 0, 255]

print(arr)
```
