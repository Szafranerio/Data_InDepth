29. How to round away from zero a float array ? (★☆☆)

https://numpy.org/doc/2.2/reference/generated/numpy.floor.html

https://numpy.org/doc/2.2/reference/generated/numpy.ceil.html

```python
#Floor
x = np.linspace(0,10, 15)
print(x)
x = np.floor(x)
print(x)
print('\n')
#Ceil
y = np.linspace(0,10,15)
print(y)
y = np.ceil(y)
print(y)
```

30. How to find common values between two arrays? (★☆☆)

```python
x = np.array([1,2,3,4,5,6,7])
y = np.array([2,4,6,8,9])

print(np.intersect1d(x,y))

#For more than two arrayes use
from functools import reduce
z = np.array([3,6,9])
reduce(np.intersect1d,(x,y,z))
```

33. How to get the dates of yesterday, today and tomorrow? (★☆☆)

```python
yesterday = np.datetime64('today', 'D') - np.timedelta64(1, 'D')

today = np.datetime64('today', 'D')

tomorrow = np.datetime64('today', 'D') + np.timedelta64(1, 'D')

print(f'Yesterda was {yesterday}, today is {today}, tomrrow will be {tomorrow}')
```
