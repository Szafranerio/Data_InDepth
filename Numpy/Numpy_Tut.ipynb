```python
# --------------------
# 1D Slicing Examples
# --------------------
x = np.array([10, 20, 30, 40, 50, 60, 70, 80])
print("Original 1D array:")
print(x)

print("\n1D slicing examples:")
print("Every 2nd element from start x[::2]     →", x[::2])
print("Every 2nd element from index 1 x[1::2]    →", x[1::2])
print("Reverse array x[::-1]    →", x[::-1])
print("Elements from index 2 to 6, step 2 x[2:7:2]   →", x[2:7:2])

# --------------------
# 2D Slicing Examples
# --------------------
a = np.arange(36).reshape(6, 6)
print("\nOriginal 2D array:")
print(a)

print("\n2D slicing examples:")
print("a[::2, ::2] → Even rows & columns:")
print(a[::2, ::2])

print("\na[1::2, ::2] → Odd rows & even columns:")
print(a[1::2, ::2])

print("\na[::-1, ::-1] → Full reversed matrix:")
print(a[::-1, ::-1])
```

```python
import numpy as np

arr = np.arange(-5, 6)
print("Array:", arr)

print("\n--- Mathematical functions ---")
print("sqrt     :", np.sqrt(np.abs(arr)))      # square root (abs to avoid negatives)
print("square   :", np.square(arr))            # square
print("power^3  :", np.power(arr, 3))          # cube
print("abs      :", np.abs(arr))               # absolute value
print("sign     :", np.sign(arr))              # sign (-1, 0, +1)

print("\n--- Exponentials & logs ---")
print("exp      :", np.exp(arr))               # e^x
print("expm1    :", np.expm1(arr))             # e^x - 1
print("log      :", np.log(np.where(arr>0, arr, np.nan)))     # ln(x), safe
print("log10    :", np.log10(np.where(arr>0, arr, np.nan)))   # base-10 log
print("log2     :", np.log2(np.where(arr>0, arr, np.nan)))    # base-2 log
```

```python
print("\n--- Trigonometric functions ---")
print("sin      :", np.sin(arr))               # sine
print("cos      :", np.cos(arr))               # cosine
print("tan      :", np.tan(arr))               # tangent
print("arcsin   :", np.arcsin(np.clip(arr/5, -1, 1)))  # inverse sine
print("arccos   :", np.arccos(np.clip(arr/5, -1, 1)))  # inverse cosine
print("arctan   :", np.arctan(arr))            # inverse tangent
print("deg2rad  :", np.deg2rad(arr))           # degrees → radians
print("rad2deg  :", np.rad2deg(arr))           # radians → degrees
```

```python
print("\n--- Rounding functions ---")
print("round    :", np.round(arr/2.5))         # round to nearest
print("floor    :", np.floor(arr/2.5))         # round down
print("ceil     :", np.ceil(arr/2.5))          # round up
print("trunc    :", np.trunc(arr/2.5))         # truncate towards 0

print("\n--- Statistics & aggregates ---")
print("sum      :", np.sum(arr))               # sum
print("mean     :", np.mean(arr))              # average
print("median   :", np.median(arr))            # median
print("std      :", np.std(arr))               # standard deviation
print("var      :", np.var(arr))               # variance
print("min      :", np.min(arr))               # min
print("max      :", np.max(arr))               # max
print("cumsum   :", np.cumsum(arr))            # cumulative sum
print("cumprod  :", np.cumprod(arr))           # cumulative product
print("prod     :", np.prod(arr))              # product of all elements

print("\n--- Array utilities ---")
print("unique   :", np.unique(arr))            # unique values
print("sort     :", np.sort(arr[::-1]))        # sorted
print("clip     :", np.clip(arr, -2, 2))       # keep within [-2, 2]
```
