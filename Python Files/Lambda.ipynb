Lambda, Map, Filter

```python
def square(num):
    return num ** 2

my_num = [1,2,3,4,5,6]
map(square,my_num)
```

```python
for item in map(square,my_num):
    print(item)
```

```python
list(map(square,my_num))
```

```python
def slicer(mystring):
    if len(mystring)%2 == 0:
        return 'EVEN'
    else:
        return 'ODD'

names = ['Adam', 'Sally', 'Thomas', 'Eve']
list(map(slicer,names))
```

```python
#Filter

def check_even(num):
    return num%2==0

my_nums = [1,2,3,4,5,6,7,8]

list(filter(check_even, my_nums))
```

```python
#Lambda

def square(num):
    result = num**2
    return result
```

```python
lambda num: num**2
```

```python
#FAST FUNCTION
list(map(lambda num: num**2, my_nums))
```

```python
list(filter(lambda num:num%2 == 0, my_nums))
```

```python
#Switch the style
list(map(lambda x:x[::-1], names))
```

```python
x = int(input('What is your number: '))
print(f"Oooops, its the square of your number {list(map(lambda num: num**2,[x]))}")
```

```python
x = int(input('What is your number: '))
print(f'Plus 15 to your number is: {list(map(lambda num: num + 15, [x]))}')
```
