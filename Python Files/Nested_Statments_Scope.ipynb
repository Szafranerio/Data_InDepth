# LEGB Rule Summary

## Local
- Variables defined inside a function are in the local scope.  
- These variables are accessible only within that function.

## Enclosing
- Variables in the scope of an enclosing function are accessible to nested functions.  
- If a variable isn’t found locally, Python looks in the enclosing scope.

## Global
- Variables defined at the top level of a script or module are in the global scope.  
- These variables can be accessed from any function unless shadowed by a local or enclosing variable.

## Built-in
- Python’s built-in names (e.g., `len`, `print`) are always accessible.  
- If a variable isn’t found in local, enclosing, or global scope, Python checks the built-in scope.

```python
#Local -> Enclossing -> Global
#Global
name = 'This is string'

def greet():
    #Enclosing
    name = 'Tom'
    def hello():
        #Local
        print('Hello '+name)
    hello()
greet()

name = 'This is string'

def greet():
    #name = 'Tom'
    def hello():
        print('Hello '+name)
    hello()
greet()
```

```python
def ran_check(num, low, high):
    if num in range(low, high + 1):
        print(f'{num} is in range of {low} and {high}')
    else:
        print('Not in range')
ran_check(5,2,7)
```
