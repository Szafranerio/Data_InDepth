```python
data.TotalFat.fillna(value=0, inplace=True)
data.head()
```

```python
data.loc[1, 'TotalFat'] = np.NaN
data.loc[1, 'Saturated Fat'] = np.NaN
data.loc[2, 'Saturated Fat'] = np.NaN
data.loc[3, 'Category'] = np.NaN
data.head()
```

Filtering of data

```python
data = pd.read_csv('./data/Mcdonalds.csv', usecols=['Item', 'Category', 'Serving Size', 'Calories', 'TotalFat'])
data.head()
```

```python
calover400 = data.Calories >= 400
calover400
```

```python
isBreakfast = data['Category'] == 'Breakfast'
isBreakfast
```

```python
#Condition plus condition
data[isBreakfast & calover400].head()
```

```python
#Prestens all data but reverse to condition
data[~ isBreakfast].head()
```

Where and query

```python
#THe values that are not filling the row, where will replace those value to NaN, it keeps the same size of the frame
data.where(calover400).head()
```

```python
data.query('Category ==  "Desserts"').head()
```

```python
data.query('Category == "Desserts" and Calories < 200')
```

```python
animals = pd.read_csv('./data/sleep_time.csv', index_col='ID').fillna(0)
animals.head()
```
