Text columns

```python
data = pd.read_csv('./data/PublicTransitExpenses.csv', usecols=['Agency', 'Reporter Type', 'Total Operating Expenses'])
data
```

```python
data['Agency'].str.contains('washington').value_counts()
```

```python
#Python is case sensetive, make it lower or upper.
data['Agency'] = data['Agency'].str.lower()
data['Agency'].str.contains('washington').value_counts()
```

```python
endswithferries = data['Agency'].str.lower().str.strip().str.endswith('ferries')
data[endswithferries]
```

```python
#Use functions to make it easier
def getComment(row):
    reporterType = row['Reporter Type']
    cost = float(row['Total Operating Expenses'].replace('$',''))
    
    if (cost > 200000):
        comment = "CLASS A"
    else:
        comment = "CLASS B"
    return (reporterType+'/'+comment)
data_over200K = data.apply(getComment, axis=1)
data_over200K.value_counts()
```

```python
data = pd.read_csv('./data/Fortune_500_2017.csv', usecols=['Rank', 'Title', 'Industry', 'Hqlocation', 'Sector'])
data.head()
```

```python
#Sector to upper()
data['Sector'] = data['Sector'].str.upper()
data
```

```python
comp = data['Industry'].str.strip().str.lower().str.contains('comp')
data[comp]
```

```python
data[['City', 'Stale']] = data['Hqlocation'].str.split(",", expand=True)
data
```

```python
def BuildShortcut(row):
    industry = row['Industry']
    result = " "
    for i in industry.split(" "):
        result += i[0]
    return result
BuildShortcut({'Industry': 'Factory Under Newspaper'})
```

```python
data['IndustryShortcut'] = data.apply(BuildShortcut, axis=1)
data.head()
```
