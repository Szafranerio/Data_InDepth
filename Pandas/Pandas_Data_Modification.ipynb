Data Modification

```python
data = pd.read_csv('./data/insurance.csv')
data.head()
```

```python
#Direct change of data
data.loc[2, 'Claims'] = 19
data.head()
```

```python
#Direct filktering to get a specific condition
isYounger = data['Age'] == '<25'
data.where(isYounger).dropna()
```

```python
data.loc[isYounger, 'Holders'].head()
```

```python
#Adding 100 to the specific value with condition
data.loc[isYounger, 'Holders'] = data.loc[isYounger, 'Holders'] + 100
data.head()
```

```python
#Multiply inome by 1.5 for scientists
profession.loc[isScientist, 'income'] = profession['income'] * 1.5
profession
```

Adding and deleting rows or columns

```python
data = pd.read_csv('./data/sleep_time.csv')
data
```

```python
#Delete columns
data.drop(columns='awake', inplace=True)
data
```

```python
#Delete rows, labels = index
data.drop(axis=0, labels= [1,2], inplace=True)
data
```

```python
#Axis = 0, always
data.drop(labels=[4], inplace=True)
data
```

```python
#The easiet and fastest version, axis is alwasy set to 0 and labels in second in documentation
data.drop(5, inplace=True)
data
```

CONCAT

```python
#Drop columns for less code and just example for concat
data.drop(columns=['sleep_rem', 'sleep_cycle', 'brainwt', 'bodywt'], axis=1, inplace=True)
```

```python
#Rememebr to add name of var before pd.concat to add a rows
new_data = {
    'name': ['Panda4'], 
    'genus': [np.NaN], 
    'vore': [np.NaN], 
    'order': [np.NaN], 
    'conservation': [np.NaN], 
    'sleep_total': [10]
}
data = pd.concat([data, pd.DataFrame(new_data)], ignore_index=True)
data
```

```python
(data['name'] == 'Panda4').value_counts()
```

```python
#Take some part of dataframe and add it by using concat
subset = data[:20]
data = pd.concat([data, subset], ignore_index=True)
data
```
