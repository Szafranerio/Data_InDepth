Columns renaming

```python
data = pd.read_csv('./data/mammals.csv')
data.head()
```

```python
#Fast and for all
data.columns = ['name', 'weight_body_kg', 'weight_brain_kg']
data.head()
```

```python
#For choosen columns
columns = {'weight_body_kg': 'weight_bodyKG', 'weight_brain_kg': 'weight_brainKG'}
data.rename(columns = columns, inplace=True)
data.head()
```

```python
data = pd.read_csv('./data/mammals.csv', index_col='name')
data.head()
```

```python
#Chagne the value in index
data.rename(index= {'Cow': 'Land Cow'}, inplace=True)
data.head()
```

```python
#Using a copy to change a value inside of dataframe. Use .loc
dataCopy = data.copy()
dataCopy.loc['Arctic fox', 'body'] = 5
dataCopy.head()
```
