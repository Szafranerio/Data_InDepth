stack, unstack - pivot a column or row to the oposit axis.

```python
data = pd.read_csv('./data/Canadian Railway Crossing Incidents.csv')
data.set_index(['Region', 'EventType'], inplace=True)
data.sort_index(inplace=True)
data
```

```python
#Select a particular row
stacked_data.loc[('Alberta','Accidents','Public passive'),0]
```

Unstack(level = -1) "-1" indicates that most right index wold be moved to the columns, level = 0 would move the most left. List slicing idea. Also reduces one indexfrom multiindex. For example from 3 to 2

```python
stacked_data.unstack(0)
```

Pivot() - group unique values within one or more discrate categories

```python
data = {"value": range(12),
        "variable": ['A']*3 + ['B']*3 + ["C"]*3 + ['D']*3,
        "date": pd.to_datetime(['2024-01-09', '2024-01-10', '2024-01-11']*4)
        }
data = pd.DataFrame(data)
data
```

```python
pivoted_data = data.pivot(index='date', columns='variable', values='value')
pivoted_data
```

```python
data = pd.read_csv('./data/marathon_results_2016.csv', usecols=['Bib', '40K', 'Half', 'Pace', 'Age', 'M/F', 'Country', 'State', 'City'], index_col='Bib')
data
```

```python
data.info()
```

```python
from datetime import datetime
from datetime import timedelta
import time
```

```python
data['40K'] = pd.to_timedelta(data['40K'], errors='coerce')
data['Half'] = pd.to_timedelta(data['Half'], errors='coerce')

data['TotalSeconds'] = data['40K'].apply(lambda x: x.total_seconds() if pd.notnull(x) else None)
data['HalfSeconds'] = data['Half'].apply(lambda x: x.total_seconds() if pd.notnull(x) else None)

data
```

```python
data_pivot_table = data.pivot_table(index='Age', columns='M/F', values='TotalSeconds', aggfunc='mean').head()
data_pivot_table
```

```python
data_pivot_table = data.pivot_table(index='Age', columns='M/F', values=['TotalSeconds', 'HalfSeconds'], aggfunc='mean').head()
data_pivot_table
```

```python
data_pivot_table.swaplevel(axis=1).head()
```

```python
data_pivot_table.swaplevel(axis=1).head().sort_index(axis=1).head()
```

Melt() 

```python
data = pd.read_csv('./data/WA_Sales_Products_2012-14.csv')
data_pivot = data.pivot_table(index='Retailer country', columns='Year', values='Revenue', aggfunc='sum')
data_pivot
```

```python
data_pivot.reset_index(inplace=True)
data_pivot.fillna(0, inplace=True)
```

```python
data_pivot.melt(id_vars='Retailer country', value_name='RevSum', var_name='Year_of_Transaction').head()
```

```python
#Unstack method, first make a pivot table
data_pivot = data.pivot_table(index='Retailer country', columns='Year', values='Revenue', aggfunc='sum')
```

```python
data = data_pivot.unstack().to_frame().swaplevel().head()
data.columns=['Rev']
data.head()
```
