Merge()

```python
cat1 = pd.read_csv('./data/northwind-mongo-master/categories_del_1.csv', usecols =["CategoryID", "CategoryName"])
cat2 = pd.read_csv('./data/northwind-mongo-master/categories_del_2.csv', usecols =["CategoryID", "CategoryName"])
```

```python
#Inner
cat1.merge(cat2)
```

```python
merged = cat1.merge(cat2, on='CategoryID', suffixes=["_1", "_2"], how='outer', indicator=True)
```

```python
filter = merged["_merge"] == "both"
merged[filter]
```

```python
#how=left
merged = cat1.merge(cat2, on='CategoryID', suffixes=["_1", "_2"], how='left', indicator=True)
merged
```

```python
#Challange
f_2015 = pd.read_csv('./data/Fortune_500_2015.csv', usecols=['rank', 'company'])
f_2016 = pd.read_csv('./data/Fortune_500_2016.csv', usecols=['rank', 'company'])
```

```python
merged = f_2015.merge(f_2016, how='outer', indicator=True, on='company', suffixes=["_2015", "_2016"])
merged
```

```python
only_2015 = merged[merged["_merge"] == 'left_only']
only_2015
```

```python
only_2016 = merged[merged["_merge"] == 'right_only']
only_2016
```

```python
#Challange
suppliers = pd.read_csv('./data/northwind-mongo-master/suppliers_del_1.csv', usecols=['SupplierID', 'CompanyName', 'City'])
products = pd.read_csv('./data/northwind-mongo-master/products_del_1.csv', usecols=['ProductID', 'ProductName', 'SupplierID'])
```

```python
suppliers
```

```python
#Inner merge
prod_and_sup = products.merge(suppliers, how="inner", on='SupplierID')
prod_and_sup
```

```python
filter = prod_and_sup_outer['_merge'] == 'left_only'
prod_and_sup_outer[filter]
```
