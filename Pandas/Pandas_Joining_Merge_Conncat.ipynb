Concat - is looking for same columns, so it can glue those dataframes. It also allows to add more then 1 dataframe, you can add 2,3,4,.... dataframes. Use ingore_index to create new idnex 

```python
suppliers = pd.read_csv('./data/northwind-mongo-master/suppliers.csv', usecols=['SupplierID', 'CompanyName', 'Country'])
suppliers.head()
```

```python
customers = pd.read_csv('./data/northwind-mongo-master/customers.csv', usecols=['CustomerID', 'CompanyName', 'Country'])
customers.head()
```

```python
#Customers are added to suppliers
suppliers_cusomter = pd.concat([suppliers, customers], ignore_index=True)
suppliers_cusomter
```

```python
prod_1 = pd.read_csv('./data/northwind-mongo-master/products.csv', usecols=['ProductID', 'ProductName'], index_col='ProductID')
prod_2 = pd.read_csv('./data/northwind-mongo-master/products.csv', usecols=['ProductID', 'UnitPrice', 'UnitsInStock'], index_col='ProductID')
```

```python
#Join data and make a value columns whcih calcualtes value of each product
joined_data = prod_1.join(prod_2).head()
joined_data['Value'] = joined_data['UnitPrice'] * joined_data['UnitsInStock']
joined_data
```
