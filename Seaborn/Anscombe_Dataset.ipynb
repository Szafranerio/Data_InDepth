```python
import seaborn as sns
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

data = sns.load_dataset("anscombe")
data.head(15)
```

```python
#How many datasets
data.dataset.value_counts()
```

```python
#Stats shows that all data statistics are same
data.groupby('dataset').agg(['mean', 'var', 'std']).round()
```

```python
sns.lmplot(data, x='x', y='y', hue='dataset', col='dataset', col_wrap=2, palette='muted');
```

The Anscombes quartet shows that datasets with identical summary statistics (mean, variance, correlation, regression line) can look completely different when visualized.

This highlights the importance of always plotting your data before drawing conclusions.
