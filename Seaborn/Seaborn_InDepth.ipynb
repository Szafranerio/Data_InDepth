```python
import pandas as pd
import seaborn as sns

data = {
    "Student": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Helen", "Ian", "Jack"],
    "Gender": ["F", "M", "M", "M", "F", "M", "F", "F", "M", "M"],
    "Age": [20, 22, 21, 23, 20, 24, 22, 21, 23, 22],
    "Study_Hours": [15, 10, 12, 9, 16, 8, 14, 13, 7, 11],
    "Score": [10, 70, 78, 65, 90, 60, 88, 82, 55, 72],
    "Course": ["Math", "Physics", "Math", "History", "Math", "History", "Physics", "History", "Math", "Physics"],
    "Exam_Date": pd.date_range("2023-05-01", periods=10, freq="D")
}

df = pd.DataFrame(data)
print(df)
```

Distribution Plots

ðŸ”¹ Histogram (sns.histplot)

Shows counts (or probability) in bins.

Good for seeing shape of data (normal, skewed, multimodal).

Example: Exam score distribution.

Use when:

You want to check frequency distribution of one numeric variable.

```python
#Histplot
sns.histplot(data=df.Score)
```

```python
sns.histplot(data=df, y=df.Score)
```

Add a kernel density estimate to smooth the histogram, providing complementary information about the shape of the distribution

```python
sns.histplot(data=df.Score, kde=True, bins=10)
```

```python
sns.histplot(data=df, x=df.Score, bins=10, hue=df.Age)
```

ðŸ”¹ Boxplot (sns.boxplot)

Summarizes data: median, quartiles, and outliers.

Compact comparison between categories.

Use when:

You want to compare distributions across groups.

You care about outliers and spread.

```python
sns.boxplot(data=df, x=df.Gender, y=df.Score)
```

ðŸ”¹ Violin Plot (sns.violinplot)

Combination of boxplot + KDE.

Shows distribution shape + summary stats.

Use when:

You want both summary stats and distribution shape per category.

```python
sns.violinplot(data=df, x = df.Gender, y = df.Age)
```

Other distrubution plots

```python
#Strip plot
sns.stripplot(df, x=df.Gender, y=df.Age)
```

```python
#Swarm plot
sns.swarmplot(df, x=df.Gender, y=df.Age)
```

```python
#ECDF Plot (Empirical Cumulative Distribution Function)
sns.ecdfplot(df["Study_Hours"])
```

ðŸ”¹ Bar Plot (sns.barplot)

Shows average (with confidence interval) per category.

Use when:

You want aggregated values (mean, median) by category.

Example: Average score by course.

```python
#It shows avg!
sns.barplot(data=df, x=df.Score, y=df.Gender)
```

```python
sns.barplot(df, y=df.Study_Hours, x=df.Course)
```

```python
#You can use different estimator
sns.barplot(df, x=df.Gender, y=df.Score, estimator='sum', errorbar=None)
```

```python
sns.catplot(df, kind = 'bar', x = 'Gender', y = 'Score', col = 'Course', errorbar=None)
```

ðŸ”¹ Count Plot (sns.countplot)

Like a bar plot, but for counts of each category.

Use when:

You want to see frequency of categorical values.

Example: How many students per course.

```python
sns.countplot(df, x=df.Course, hue=df.Gender)
```

```python
sns.countplot(df, x= df.Gender)
```

```python
sns.countplot(df, x= df.Gender, hue=df.Course)
```

ðŸ”¹ Scatter Plot (sns.scatterplot)

Plots individual points.

Can show correlation, clusters, outliers.

Use when:

You want to see relationship between 2 variables.

Example: Study hours vs. score.

```python
sns.scatterplot(df, x=df.Study_Hours, y=df.Score)
```

```python
sns.scatterplot(df, x=df.Study_Hours, y=df.Gender)
```

```python
markers = {'M': 'o', 'F': 'v'}
sns.scatterplot(df, x=df.Study_Hours, y=df.Score, style=df.Gender, markers=markers)
```

ðŸ”¹ Line Plot (sns.lineplot)

Shows trend of variable(s) across time or ordered axis.

Use when:

You want to show trends or time series data.

Example: Scores over exam dates.

```python
import matplotlib.pyplot as plt
plot = sns.lineplot(df, x=df.Exam_Date, y=df.Score)
plt.xticks(rotation = 30)
plt.show()
```

ðŸ”¹ Clustermap (sns.clustermap)

Heatmap + hierarchical clustering.

Groups similar rows/columns together.

Use when:

You want to see patterns and clusters in your matrix data.

Example: Students grouped by similar study/scores

```python
import numpy as np
np.random.seed(42)
df = pd.DataFrame({
    "Course": np.repeat(["Math", "Physics", "History"], 200),
    "Score": np.concatenate([
        np.random.normal(75, 10, 200),   # Math
        np.random.normal(70, 12, 200),   # Physics
        np.random.normal(65, 15, 200)    # History
    ])
})

df["Course"] = pd.Categorical(df["Course"], categories=["History", "Physics", "Math"], ordered=True)

g = sns.FacetGrid(df, row="Course", hue="Course", aspect=4, height=1.5, palette="Set2")

g.map(sns.kdeplot, "Score", fill=True, alpha=0.7, linewidth=1.5)
g.map(plt.axhline, y=0, lw=2, clip_on=False)
g.add_legend()
g.set_titles("")
g.set(yticks=[], ylabel="")
g.despine(bottom=True, left=True)

plt.show()
```
