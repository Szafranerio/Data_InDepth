```python
import pandas as pd
import seaborn as sns

data = {
    "Student": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Helen", "Ian", "Jack"],
    "Gender": ["F", "M", "M", "M", "F", "M", "F", "F", "M", "M"],
    "Age": [20, 22, 21, 23, 20, 24, 22, 21, 23, 22],
    "Study_Hours": [15, 10, 12, 9, 16, 8, 14, 13, 7, 11],
    "Score": [10, 70, 78, 65, 90, 60, 88, 82, 55, 72],
    "Course": ["Math", "Physics", "Math", "History", "Math", "History", "Physics", "History", "Math", "Physics"],
    "Exam_Date": pd.date_range("2023-05-01", periods=10, freq="D")
}

df = pd.DataFrame(data)
print(df)
```

Distribution Plots

ðŸ”¹ Histogram (sns.histplot)

Shows counts (or probability) in bins.

Good for seeing shape of data (normal, skewed, multimodal).

Example: Exam score distribution.

Use when:

You want to check frequency distribution of one numeric variable.

```python
#Histplot
sns.histplot(data=df.Score)
```

```python
sns.histplot(data=df, y=df.Score)
```

Add a kernel density estimate to smooth the histogram, providing complementary information about the shape of the distribution

```python
sns.histplot(data=df.Score, kde=True, bins=10)
```

```python
sns.histplot(data=df, x=df.Score, bins=10, hue=df.Age)
```

ðŸ”¹ Boxplot (sns.boxplot)

Summarizes data: median, quartiles, and outliers.

Compact comparison between categories.

Use when:

You want to compare distributions across groups.

You care about outliers and spread.

```python
sns.boxplot(data=df, x=df.Gender, y=df.Score)
```

ðŸ”¹ Violin Plot (sns.violinplot)

Combination of boxplot + KDE.

Shows distribution shape + summary stats.

Use when:

You want both summary stats and distribution shape per category.

```python
sns.violinplot(data=df, x = df.Gender, y = df.Age)
```

Other distrubution plots

```python
#Strip plot
sns.stripplot(df, x=df.Gender, y=df.Age)
```

```python
#Swarm plot
sns.swarmplot(df, x=df.Gender, y=df.Age)
```

```python
#ECDF Plot (Empirical Cumulative Distribution Function)
sns.ecdfplot(df["Study_Hours"])
```
