```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data/Crime_Data_from_2020_to_Present.csv')
data
```

```python
data.columns
```

```python
to_drop = ['DR_NO', 'TIME OCC', 'AREA',
       'Rpt Dist No', 'Part 1-2', 'Crm Cd', 'Crm Cd Desc', 'Mocodes', 'Vict Descent', 'Premis Cd', 'Premis Desc',
       'Weapon Used Cd', 'Weapon Desc', 'Status', 'Status Desc', 'Crm Cd 1',
       'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street']
```

```python
data = data.drop(columns=to_drop, axis=1)
data.head()
```

```python
data.info()
```

```python
#Check null values
sns.heatmap(data.isnull())
```

```python
print(f"Oldest person: {data['Vict Age'].max()}")
print(f"Youngest person: {data['Vict Age'].min()}")
```

```python
data = data.rename(columns={'Vict Age': 'Vict_Age'})
data = data.drop(data[data.Vict_Age < 1].index)
data
```

```python
plt.figure(figsize=(12,10))
sns.histplot(data, x = 'Vict_Age', kde=True)
plt.xticks(rotation = 90)
plt.show()
```

```python
#Stats per vict sex
data = data.rename(columns={'Vict Sex': 'Vict_Sex'})
```

```python
sns.countplot(data, x = data.Vict_Sex)
```

```python
data.Vict_Sex = data.Vict_Sex.map({'X': 3, 'H': 2, 'M':1, 'F':0})
data.head()
```

```python
data = data.drop(data[data.Vict_Sex > 1].index)
data.Vict_Sex = data.Vict_Sex.map({1: 'M', 0: 'F'})
sns.countplot(data, x = data.Vict_Sex)
plt.show()
```
